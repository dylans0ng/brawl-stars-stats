```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
page = requests.get('https://www.brawlplanet.com/brawlers')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
tables = soup.find_all('tbody', attrs={'class': 'divide-y divide-gray-200 dark:divide-gray-800'})
brawler_percentage_table = tables[0]
brawler_info_table = tables[1]
```

```python
brawler_percentage_list = []

for row in brawler_percentage_table.find_all('tr'):
    brawler_name = row.find_all('td')[0].text
    win_rate = row.find_all('td')[1].text
    use_rate = row.find_all('td')[2].text
    starplayer_rate = row.find_all('td')[3].text

    percentage_per_brawler = []

    percentage_per_brawler.append(brawler_name)
    percentage_per_brawler.append(win_rate)
    percentage_per_brawler.append(use_rate)
    percentage_per_brawler.append(starplayer_rate)

    brawler_percentage_list.append(percentage_per_brawler)

brawler_percentage_df = pd.DataFrame(brawler_percentage_list, columns=['Brawler Name', 'Win Rate (%)', 'Use Rate (%)', 'Starplayer Rate (%)'])
brawler_percentage_df
```

```python
brawler_info_list = []

for row in brawler_info_table.find_all('tr'):
    brawler_name = row.find_all('td')[0].text
    health = row.find_all('td')[1].text
    damage = row.find_all('td')[2].text
    attack_range = row.find_all('td')[3].text
    reload_speed = row.find_all('td')[4].text
    movement_speed = row.find_all('td')[5].text

    info_per_brawler = []

    info_per_brawler.append(brawler_name)
    info_per_brawler.append(health)
    info_per_brawler.append(damage)
    info_per_brawler.append(attack_range)
    info_per_brawler.append(reload_speed)
    info_per_brawler.append(movement_speed)

    brawler_info_list.append(info_per_brawler)

brawler_info_df = pd.DataFrame(brawler_info_list, columns=['Brawler Name', 'Level 1 Health', 'Level 1 Damage', 'Attack Range', 'Reload Speed', 'Movement Speed'])
brawler_info_df
```

```python
df = pd.merge(brawler_percentage_df, brawler_info_df, on='Brawler Name')
df
```

```python
df.to_csv('Brawl-Stars-Stats.csv', index=False)
```
